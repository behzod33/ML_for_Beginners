```python
import numpy as np
import pandas as pd

from sklearn.datasets import make_regression
```

```python
def generate_sklearn_data(n_samples=200, n_features=5, noise=10, random_state=42):
    """Генерация данных через sklearn"""
    X, y, true_coef = make_regression(
        n_samples=n_samples, 
        n_features=n_features, 
        noise=noise, 
        coef=True, 
        random_state=random_state,
        bias=3.0
    )
    return pd.DataFrame(X, columns=[f'f{i}' for i in range(n_features)]), pd.Series(y), true_coef
```

```python
X, y, true_coef = generate_sklearn_data()

n_iter = 10
learning_rate = 0.5

true_coef
```

```python
def mse_func(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

def grad_func(X, y_true, y_pred):
    n_samples = X.shape[1]
    return (2 / n_samples) * X.T @ (y_pred - y_true)

X_with_bias = np.concatenate([np.ones(X.shape[0]).reshape(-1,1), X], axis = 1)
num_samples, num_features = X.shape

weights = np.ones(X_with_bias.shape[1])
y_pred = np.full(X_with_bias.shape[0], np.nan)

for iter in range(n_iter):
    y_pred =  X_with_bias @ weights
    mse = mse_func(y, y_pred)
    weights = weights - learning_rate * grad_func(X_with_bias, y, y_pred)
    print(mse)
```

```python
class MyLineReg():
    def __init__(self, n_iter: int = 10, learning_rate: float = 0.5, weights=None):
        self.n_iter = n_iter
        self.learning_rate = learning_rate
        
    def __str__(self):
        return f'MyLineReg class: n_iter={self.n_iter}, learning_rate={self.learning_rate}'
    
    def fit(self, X: pd.DataFrame, y: pd.Series, verbose=False):
        self.X = np.concatenate([np.ones(X.shape[0]).reshape(-1,1), X], axis = 1)
        self.y = y
        
        num_samples, num_features = self.X.shape
        self.weights = np.ones(num_features)

        for iter in range(self.n_iter):
            self.y_pred =  self.X @ self.weights
            mse = ((self.y - self.y_pred) ** 2).mean()
            dw = (2 / num_samples) * self.X.T @ (self.y_pred - self.y)
            self.weights = self.weights - self.learning_rate * dw

            if (verbose != False) and (verbose % iter == 0):
                print(f'{iter} | loss: {mse}')
                
    def get_coef(self):
        return self.weights[1:]
```

```python
lr = MyLineReg()

lr.fit(X, y)

model_coef = lr.get_coef().mean()

print(f'Model Coef: {model_coef}')
print(f'True Coef: {true_coef.mean()}')
```
